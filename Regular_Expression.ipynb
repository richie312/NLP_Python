## Stemming
User defined function for stemming the words.


```python
suffix = ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment','er','ion']

def stem(word):
    for string in suffix:
        if word.endswith(string):
            return word[:-len(string)]
    return word
text="unindent does not match any outer indentation level"
list_of_words = text.split()

stemmed_words = [stem(v) for v in list_of_words]

print("{}".format(stemmed_words))
```

## Different types of expression with its search result and class type of result

```python
import re
import nltk
exp = '[a-zA-Z]+','[A-Z][a-z]*','p[aeiou]{,2}t','\d+(\.\d+)?','([^aeiou][aeiou][^aeiou])*','\w+|[^\w\s]+'
word = "Processing"
print('{0:15}{1:20}'.format('Expression','Search Result'))
for i in range(len(exp)):
    print(exp[i],re.findall(r"{}".format(exp[i]),word))
for j in range(len(exp)):     
    print(nltk.re_show(exp[j],word))

```
